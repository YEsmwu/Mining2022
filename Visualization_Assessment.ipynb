체중과 총콜레스테롤은 어떠한 상관관계를 가지고 있을까? 전문가에 의하면 과체중이나 비만은 나쁜 콜레스테롤의 수치를 올린다고 한다. 하지만 위의 도표로 보았을 때에는 체중과 총콜레스테롤은 큰 상관관계를 가지고 있지 않는 것으로 보인다. 그 이유는 체중이 많이 나가는 사람들이라고 꼭 총콜레스트롤의 수치가 높게 나오지 않고, 체중에 비해서 콜레스테롤의 수치가 높게 나오는 사람도 분명히 존재하기 때문이다.

```python
smoke = health['Age(unit.5)'] >= 20
print (smoke)
print(health[smoke])

df_health = health.groupby('Smoke', as_index = False).agg(n = ('Smoke' , 'count'))

df_health
```

```python
sns.barplot(data = df_health, x = 'Smoke', y = 'n')
```

흡연 비율을 살펴보기 위해서 막대그래프를 통해서 흡연자의 비율을 살펴보았다. 법적으로 흡연이 가능한 연령대(20세 이상) 중 흡연자와 비흡연자, 현재는 피고 있지 않은 사람을 분석해 보았다. 분석 결과, 흡연자의 비율은 그렇게 높지 않은 것으로 보인다. 흡연자의 비율은 비흡연자의 약 1/3을 차지한다.
